```python
# First of all lets import all necessary libraries

import numpy as np
import pandas as pd
```

```python
# Now we can start by reading the datasets
# We will concatenate all csv files into one single dataset

filenames = ['season-0910.csv', 
             'season-1011.csv', 
             'season-1112.csv', 
             'season-1213.csv', 
             'season-1314.csv', 
             'season-1415.csv', 
             'season-1516.csv', 
             'season-1617.csv', 
             'season-1718.csv', 
             'season-1819.csv']
datasets = [pd.read_csv(f) for f in filenames]
data = pd.concat(datasets, axis=0, ignore_index=True)
```

```python
# Our dataset looks like this

data
```

```python
# We can start doing preprocessing
# Let's look at colums

data.info()
```

```python
# Date column (that is string) can be converted to actual Date format 

data['Date'] = pd.to_datetime(data['Date'])
```

```python
# Div column's values are same(SP1) for all records

data['Div'].value_counts()
```

```python
# So it can be deleted without consequences

data = data.drop(['Div'], axis=1)
```

```python
# There is columns with too many null values
# They can be removed too

data = data.drop(['GBH', 'GBD', 'GBA', 'SBH',
                  'SBD', 'SBA', 'SJH', 'SJD',
                  'SJA', 'BSH', 'BSD', 'BSA',
                  'PSCH', 'PSCD', 'PSCA'], axis=1)
```

```python
# We can also delete some columns to decrease their amount for simplicity of predicting
# The actual reason I am deleting these columns is it is
# hard to find the odds of all provided betting odds

data = data.drop(['Bb1X2', 'BbMxH', 'BbAvH', 'BbMxD', 'BbAvD', 'BbMxA', 'BbAvA',
       'BbOU', 'BbMx>2.5', 'BbAv>2.5', 'BbMx<2.5', 'BbAv<2.5', 'BbAH', 'BbAHh',
       'BbMxAHH', 'BbAvAHH', 'BbMxAHA', 'BbAvAHA'], axis=1)
```

```python
# Let's take a look for our dataset

data.isnull().any()
```

```python
# We still have null values, but their amount is small
# They can be replaced with average values

data = data.fillna(data.mean())
```

```python
# Last thing we must implement is to normalize categorical column values of type string

#data['HomeTeam'] = data['HomeTeam'].astype('category')
#data['AwayTeam'] = data['AwayTeam'].astype('category')

#cat_columns = data.select_dtypes(['category']).columns

#data[cat_columns] = data[cat_columns].apply(lambda x: x.cat.codes)

from sklearn.preprocessing import LabelEncoder
labelencoder= LabelEncoder()
data['HomeTeam'] = labelencoder.fit_transform(data['HomeTeam'])
data['AwayTeam'] = labelencoder.fit_transform(data['AwayTeam'])
```

```python
# Preprocessing of dataset is finished
# Now we can begin to make predictions
# Before get started, let's look at our remaining columns

data.columns
```

```python
# ['Date', 'HomeTeam', 'AwayTeam'] - basic match info(independent data)

# ['B365H', 'B365D', 'B365A', 'BWH',
# 'BWD', 'BWA', 'IWH', 'IWD', 'IWA',
# 'LBH', 'LBD', 'LBA', 'WHH', 'WHD',
# 'WHA', 'VCH', 'VCD', 'VCA', 'PSH', 
# 'PSD', 'PSA'] - bookmakers betting odds(independent data)

# ['FTHG', 'FTAG', 'FTR', 'HTHG', 'HTAG', 'HTR', 'HS', 
# 'AS', 'HST', 'AST', 'HF', 'AF', 'HC', 
# 'AC', 'HY', 'AY', 'HR', 'AR'] - statistical information after match is played(dependent data)
```

```python
# Every our ML algorithms requires train and test part of given dataset
# Let's split it

from sklearn.model_selection import train_test_split

x = data[['HomeTeam', 'AwayTeam', 
          'B365H', 'B365D', 'B365A', 
          'BWH', 'BWD', 'BWA', 
          'IWH', 'IWD', 'IWA', 
          'LBH', 'LBD', 'LBA', 
          'WHH', 'WHD', 'WHA', 
          'VCH', 'VCD', 'VCA', 
          'PSH', 'PSD', 'PSA']]
y = data['FTR']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=9)
```

#### Predicting FTR with Naive Bayes

```python
# I am going to predict 2 games (FCBarcelona - Real Madrid)
# FCBarcelona -[0:0]- Real Madrid (2019-12-18)
# Real Madrid -[2:0]- FCBarcelona (2020-03-01)
# Note that our dataset does not contain these games

# At this stage let's find out numerical values of these clubs
# FCBarcelona's code is 4
# Real Madrid's code is 23

data
```

```python
# Now we can generate a Naive Bayes model from our data set

from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
# model.fit(x_train, y_train)
model.fit(x, y)
```

```python
# It scores only

model.score(x_test, y_test)
```

```python
# FCBarcelona -[0:0]- Real Madrid (2019-12-18, season 2019/2020, Camp Nou)
# https://chucknorisk.com/matches/fc-barcelona-vs-real-madrid-2019-12-18/

model.predict([[4, 23,
                1.65, 4.20, 4.75,
                1.72, 4.20, 4.33,
                1.75, 4.00, 4.30,
                1.67, 4.00, 4.50,
                1.70, 4.20, 4.50,
                0, 0, 0,
                1.71, 4.12, 4.87]])
```

```python
# Real Madrid -[2:0]- FCBarcelona (2020-03-01, season 2019/2020, Santiago Bernabeu)
# https://chucknorisk.com/matches/real-madrid-vs-fc-barcelona-2020-03-01/

model.predict([[23, 4,
                2.25, 3.60, 3.10,
                2.20, 3.60, 3.10,
                2.30, 3.60, 2.95,
                2.25, 3.50, 2.90,
                0, 0, 0,
                0, 0, 0,
                2.35, 3.58, 3.05]])
```

As we can see from these two predictions Naive Bayes Classifier model trained by last 10 year games successfully predicted FTR(Full Time Result) of FCBarcelona and Real Madrid's games, which were played only after the dataset was introduced.
